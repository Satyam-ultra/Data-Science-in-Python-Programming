# **Objective:**

1.   Creating a list
2.   Creating a dictionary
3.   Creating a set





```python
# Creating my list
my_list = [1,3,5,7]

# Adding an element to my list
my_list.append(9)

#Removing an element from my list
my_list.remove(5)

# Modifying a number in my list
my_list[0] = 11

print("Updated List:", my_list)
```

```python
# Creating a dictionary
my_dict = {'name': 'Satyam', 'age': '22', 'city': 'Pune', 'designation': 'Student'}

# Adding an element into my dictionary
my_dict['gender'] = 'Male'

# Removing Element
del my_dict['designation']

# Modifying Element
my_dict['name'] = 'Satyam Salunkhe'

print("Updated Dict:", my_dict)
```

```python
# Creating a set
my_set = {2,4,6,8,10}

# Adding
my_set.add(12)

# Removing
my_set.remove(6)

# Modifying
my_set.discard(10)
my_set.add(14)

print("Updated Set:", my_set)
```

## Conclusion

Successfully created a list, a dictionary, and a set. also, performed various operations on adding an element, removing an element, and modifying an element in a list, a dictionary, and a set in my 1st task.
